Task - 3 - Outlier Detection:
Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
2. Define outliers using the 1.5 × IQR rule.
3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.
4. Discuss the characteristics of these outliers and their potential impact on regional averages.

```python
#percentile 25 is Q1 whereas 75 means Q3
score_Q1=np.percentile(south_asian_df['score'],25)
score_Q3=np.percentile(south_asian_df['score'],75)
#tracing out the Inter-quartile range of the score
score_IQR=score_Q3-score_Q1
#Also tracing out upper and lower bound for outliers.
lower_bound_score=score_Q1-1.5*score_IQR
upper_bound_score=score_Q3+1.5*score_IQR

GDPperCapita_Q1=np.percentile(south_asian_df['Log GDP per capita'],25)
GDPperCapita_Q3=np.percentile(south_asian_df['Log GDP per capita'],75)

GDPperCapita_IQR=GDPperCapita_Q3-GDPperCapita_Q1

lower_bound_GDPperCapita=GDPperCapita_Q1-1.5*GDPperCapita_IQR
upper_bound_GDPperCapita=GDPperCapita_Q3+1.5*GDPperCapita_IQR

print("The lower bound for score is:",lower_bound_score)
print("The upper bound for score is:",upper_bound_score)
print("\n")

outliers=south_asian_df[
        (south_asian_df['score']<lower_bound_score)
        | (south_asian_df['score']>upper_bound_score )
        |(south_asian_df['Log GDP per capita']<lower_bound_GDPperCapita)
        |( south_asian_df['Log GDP per capita']>upper_bound_GDPperCapita)
        ]
print(outliers)

#ploting a scattered plot
plt.figure(figsize=(10,6))
sns.scatterplot(data=south_asian_df,x='Log GDP per capita',y='score',alpha=0.5,color='blue')
sns.scatterplot(data=outliers,x='Log GDP per capita',y='score',alpha=0.5,color='red')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')
plt.title('Scatter Plot of Score vs Log GDP per Capita')
plt.show()



# plt.title("Scatter Plot of Outliers")
# plt.xlabel("Log GDP per Capita")
# plt.ylabel("Score")
# plt.show()

#4
print("Characteristics of these outliers and their potential impact on regional averages is these outlier can majorly vary the data as it increases the range of the data. \n"
,"The calculated result becomes inaccurate in the presence of these outliers. ")
```

3. Metric Comparisons:

• Compare key metrics like GDP per Capita, Social Support, and Healthy Life Expectancy between the regions using grouped bar charts.

• Which metrics show the largest disparity between the two regions?

```python
# Step 1: Define regions and their respective countries
region_mapping = {
    "South Asia": ["Afghanistan", "Bangladesh", "Bhutan", "India", "Maldives", "Nepal", "Pakistan", "Sri Lanka"],
    "Middle East" : [ "Bahrain", "Iran", "Iraq", "Israel", "Jordan", "Kuwait", "Lebanon", "Oman", "Palestine", "Qatar", "Saudi Arabia", "Syria", "United Arab Emirates", "Yemen"]
}

def get_region(country):
    for region, countries in region_mapping.items():
        if country in countries:
            return region
    return "Unknown Region"

# df['Regional indicator'] = df['Country name'].apply(get_region)
middle_east_and_SouthAsian_df=df[df['Country name'].isin(middle_east_countries)|df['Country name'].isin(south_asian_countries)]
# print(middle_east_and_SouthAsian_df)
middle_east_and_SouthAsian_df['Region indicator']=middle_east_and_SouthAsian_df['Country name'].apply(get_region)
# print(middle_east_and_SouthAsian_df)

regional_means = middle_east_and_SouthAsian_df.groupby('Region indicator')[['Log GDP per capita', 'Social support', 'Healthy life expectancy']].mean()

regional_means.plot(kind="bar",figsize=(10,6),color=['green','blue','red'])
plt.title("Comparison of three main key metrics among the regions.")
plt.xlabel("Region")
plt.ylabel("Mean Value")
plt.xticks(rotation=60)
plt.legend(title="Metrics")
plt.show()
print("\n")

# Analyzing disparity
# regional_means.max() will return the maximum value for each column.
largest_disparity = regional_means.max() - regional_means.min()
print("Largest disparity observed in the following metric(s):")
print(largest_disparity[largest_disparity == largest_disparity.max()])



```
